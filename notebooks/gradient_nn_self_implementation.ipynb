```python
def cost_function(y,h):
    h=h.reshape(len(h),1)
    y=y.reshape(len(y),1)
    #print('h',h.shape)
    #print('y',y.shape)
    J=(np.linalg.norm(y-h))**2/len(y)
    return J*100

def MSE(y,h):
    h=h.reshape(len(h),1)
    y=y.reshape(len(y),1)
    #print('h',h.shape)
    #print('y',y.shape)
    MSE=sum((y-h)*(y-h))/len(y)
    return MSE*100


def sigmoid(x):
        return 1/(1+np.exp(-x))


class regressor_nn:
    
    def __init__(self, architecture):
        self.architecture = architecture
    
    def fit(self,X,y,learning_rate,minimum_cost,external_matrix=None):
        #print(y)
        iterator=0
        cost=100
        
        while (cost>=minimum_cost):
            #accumulator init. Auxilar array for calculating the gradient
            accumulator={}
            for l in np.arange(1,len(self.architecture),1):
                accumulator[l]=np.zeros((self.architecture[l],self.architecture[l-1]))

            #list for holding the prediction on traing, it will update in every iteration
            prediction_cost=[]

            #Looping the each row on the data
            for value in np.arange(0,len(X),1):
                row=X[value].reshape(len(X[value]),1)
                target=y[value].reshape(len(y[value]),1)

                # esta matrix no tiene posicion zero, wiething array matrix 
                # initial guess
                if(iterator==0):
                    if(external_matrix):
                        dicc_matrix_theta=external_matrix
                        
                    else:
                        dicc_matrix_theta={}
                        for l in np.arange(1,len(self.architecture),1):
                            #np.random.seed(seed=100)
                            #dicc_matrix_theta[l]=np.random.rand(self.architecture[l],self.architecture[l-1])/10000
                            dicc_matrix_theta[l]=matrixes_ext_o[l]

                #forward propagation    
                dicc_activation_vectors={}
                potential_vectors={}
                for l in np.arange(0,len(self.architecture),1):
                    if l==0:
                        dicc_activation_vectors[0]=row
                    elif l!=len(self.architecture)-1:
                        z=np.dot(dicc_matrix_theta.get(l),dicc_activation_vectors.get(l-1))
                        z=z+np.ones(self.architecture[l]).reshape(self.architecture[l],1)
                        potential_vectors[l]=z
                        dicc_activation_vectors[l]=sigmoid(z)
                    else:
                        z=np.dot(dicc_matrix_theta.get(l),dicc_activation_vectors.get(l-1))
                        z=z+np.ones(self.architecture[l]).reshape(self.architecture[l],1)
                        potential_vectors[l]=z
                        dicc_activation_vectors[l]=z
                        prediction_cost.append(z)

                #backward propagation
                delta_vectors={}
                for l in np.flip(np.arange(1,len(self.architecture),1)):
                    if l==len(self.architecture)-1:
                        delta_vectors[l]=2*(potential_vectors[l]-target)
                    else:
                        sigma_prime=dicc_activation_vectors.get(l)*(1-dicc_activation_vectors.get(l))
                        delta_vectors[l]=(sigma_prime)*np.dot(dicc_matrix_theta.get(l+1).T,delta_vectors.get(l+1))

                #accumulator for the gradient
                for l in np.arange(1,len(self.architecture),1):
                    accumulator[l]=accumulator.get(l)+np.dot(delta_vectors.get(l),dicc_activation_vectors.get(l-1).T)
        
            # end of loop in rows
            # check if the cost function increase, this should never happend
            
            #print(cost_function(y,np.array(prediction_cost)),iterator)
            if((cost<cost_function(y,np.array(prediction_cost))) and iterator!=0):
                print('ajust the learning rate')
                break        
            #if(iterator>15000):
                #print('not convergence')
                #break    
            cost=cost_function(y,np.array(prediction_cost))
            for l in np.arange(1,len(self.architecture),1):
                    dicc_matrix_theta[l]=dicc_matrix_theta.get(l)-learning_rate*(accumulator.get(l)/len(X))
            iterator=iterator+1
            
        self.w_opt = dicc_matrix_theta
        return dicc_matrix_theta
        
    def predict(self,X):
        h={}
        for value in np.arange(0,len(X),1):
            row=X[value].reshape(len(X[value]),1)
            #target=y[value].reshape(len(y[value]),1)
            dicc_matrix_theta=self.w_opt

            dicc_activation_vectors={}
            potential_vectors={}
            for l in np.arange(0,len(self.architecture),1):
                if l==0:
                    dicc_activation_vectors[0]=row
                elif l!=len(self.architecture)-1:
                    z=np.dot(dicc_matrix_theta.get(l),dicc_activation_vectors.get(l-1))
                    z=z+np.ones(self.architecture[l]).reshape(self.architecture[l],1)
                    potential_vectors[l]=z
                    dicc_activation_vectors[l]=sigmoid(z)
                else:
                    z=np.dot(dicc_matrix_theta.get(l),dicc_activation_vectors.get(l-1))
                    z=z+np.ones(self.architecture[l]).reshape(self.architecture[l],1)
                    potential_vectors[l]=z
                    dicc_activation_vectors[l]=z
            h[value]=float(potential_vectors.get(len(self.architecture)-1))
        h=np.array(list(h.values()))
        h=h.reshape(len(h),1)
        return(h)
        
```

```python

def Theils_U(data_navie,F_t):
    Y_t=data_navie
    numerator  =((((F_t.shift(-1)-Y_t.shift(-1))/Y_t)**2.0))
    denominator=(((Y_t.shift(-1)-Y_t)/Y_t)**2.0)
    numerator=list(numerator.dropna().values)
    denominator=list(denominator.dropna().values)
    try:
        Theils_U=np.sqrt(sum(numerator)/sum(denominator))
    except:
        Theils_U=np.sqrt(sum(numerator))
    return Theils_U

def lager(data,n):
    dic={}
    for each_n in np.arange(1,n+1,1):
        values=data.shift(each_n).values
        row_flatten=[item[0] for item in values]
        dic[each_n]=row_flatten
    return dic

def data_spliter(data,k):
    size=len(data)
    dicc={}
    current=0
    frac=int(size/k)
    #print(frac,size,k)
    for shunk in np.arange(1,k+1,1):
        #print(current,frac*shunk)
        if shunk==k:
            dicc[shunk]=data[current:]
        else:
            dicc[shunk]=data[current:frac*shunk]
        current=frac*shunk
    return dicc


def W_opt(data,target,alpha):
    m=data.shape[1]
    k=data.shape[0]
    term_1=np.linalg.pinv(np.dot(data.T,data)+ np.identity(m)*alpha*alpha)
    term_2=np.dot(data.T,target)
    return np.dot(term_2,  term_1)


def pipeline_cv(df,lags,folds,type_model,alpha=None,h_units=None,minimum_cost=None,external_matrix=None):
    df_lag=pd.DataFrame.from_dict(lager(df,lags), orient='columns')
    df_lag=df_lag.dropna()
    if(type_model=='linear_regression'):
        df_lag['bias']=1
    data_original_moved=df.shift(-lags).dropna()
    y=np.array(data_original_moved.iloc[:,0])
    if(len(list(df.columns))>1):
        data_original_moved.drop(data_original_moved.columns[[0]],axis=1,inplace=True)
        df_total=pd.concat([data_original_moved.reset_index().drop(['date'],axis=1),
                         df_lag.reset_index().drop(['index'],axis=1)], axis=1)
        X=np.array(df_total)
    else:
        X=np.array(df_lag)
    X_k_fold=data_spliter(X,folds)
    y_k_fold=data_spliter(y,folds)
    CV_train=[]
    CV_test=[]
    Theils_number=[]
    for each_fold in np.arange(1,folds+1,1):
        p_train=int(len(X_k_fold.get(each_fold))*0.75)
        train_X=X_k_fold.get(each_fold)[0:p_train]
        train_y=y_k_fold.get(each_fold)[0:p_train]
        test_X =X_k_fold.get(each_fold)[p_train:]
        test_y =y_k_fold.get(each_fold)[p_train:]
        if(type_model=='linear_regression'):
            W=W_opt(train_X,train_y,alpha)
            pred_test =np.dot(W,  test_X.T)
            pred_train=np.dot(W, train_X.T)
        
        if(type_model=='neuron_network'):
            train_y=train_y.reshape(len(train_y),1)
            #n_model=regressor_nn(architecture=[lags,math.ceil((lags+1)/2),1])
            n_model=regressor_nn(architecture=[lags,h_units,1])
            
            print(lags)
            matrix=n_model.fit(X=train_X,y=train_y,learning_rate=0.1/2,
                               minimum_cost=minimum_cost,external_matrix=external_matrix)
            pred_test=n_model.predict(test_X)
            pred_train=n_model.predict(train_X)


        MSE_function_train=MSE(train_y,pred_train)
        MSE_function_test =MSE(test_y,pred_test)
        #print('cost_function_train',cost_function_train)
        #print('cost_function_test',cost_function_test)
        
        CV_train.append(MSE_function_train)
        CV_test.append(MSE_function_test)
        df_test_y=pd.DataFrame(test_y)
        df_pred  =pd.DataFrame(pred_test)
        Theils_number.append(Theils_U(df_test_y,df_pred))

    CV_train_mean=np.mean(MSE_function_train)   
    CV_test_mean=np.mean(MSE_function_test)
    Theils_number_mean=np.mean(Theils_number)
    if(type_model=='neuron_network'):
        result=[CV_train_mean,CV_test_mean,Theils_number_mean,alpha,matrix]
    else:
        result=[CV_train_mean,CV_test_mean,Theils_number_mean,alpha]
    return result
```

```python
def cv_bias_variance_plotter(trend_parameter_optimization,lab,label_x,mode=None):
    fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(16,7))
    v_train=[]
    v_test=[]
    U_test=[]
    alpha=[]
    matrixes=[]
    contador=0
    for each_model in trend_parameter_optimization:
        if(each_model[3]=='linear_regression'):
            m=pipeline_cv(each_model[0],each_model[1],each_model[2],each_model[3],each_model[4])
        else:
            if(mode):
                if(contador==0):
                    m=pipeline_cv(df=each_model[0],lags=each_model[1],
                                  folds=each_model[2],type_model=each_model[3],
                                  h_units=each_model[4],minimum_cost=each_model[5])
                    matrix=m[4]
                else:
                    m=pipeline_cv(df=each_model[0],lags=each_model[1],
                                  folds=each_model[2],type_model=each_model[3],
                                  h_units=each_model[4],minimum_cost=each_model[5],external_matrix=matrix)
                    matrix=m[4]
                matrixes.append(matrix)
            else:
                m=pipeline_cv(df=each_model[0],lags=each_model[1],folds=each_model[2],type_model=each_model[3],h_units=each_model[4],minimum_cost=each_model[5])
            #[df_p,lag,folds,'neuron_network',None,hidden_unit]
            #df,lags,folds,type_model,h_units=None
        v_train.append(m[0])
        v_test.append(m[1])
        U_test.append(m[2])
        alpha.append(m[3])
    #ax.plot(label_x,v_train, '-o' ,  label="CV train  MSE, Minimum="+'%.3f'%min(v_train)+"% "+r"$\alpha$="+str(alpha[0]))
    #ax.plot(label_x,v_test, '-o',  label="CV Test  MSE, Minimum="+'%.3f'%min(v_test)+"% "+r"$\alpha$="+str(alpha[0]))
    #ax.plot(label_x,v_train, '-o' ,  label="CV train  MSE, Minimum="+'%.3f'%min(v_train)+"%" )
    ax.plot(label_x,v_test, '-o',  label="CV Test  MSE, Minimum="+'%.3f'%min(v_test)+"% ")
    ax2 = ax . twinx () 
    ax2.plot(label_x,U_test, 'x', label=r"Theil's statistic" ,  color="green")
    print('Minimum Theils statistic',min(U_test),U_test)
    print('CV train  MSE, Minimum=',min(v_train),v_train)
    print('CV Test  MSE, Minimum=',min(v_test),v_test)
    ax.legend(loc='best')
    ax.set_xlabel(lab ,fontdict=font)
    
    ax.set_ylabel('MSE %', fontdict=font)
    ax2.set_ylabel('Theils statistic', fontdict=font)
    ax.grid(True)
    return matrixes
```
