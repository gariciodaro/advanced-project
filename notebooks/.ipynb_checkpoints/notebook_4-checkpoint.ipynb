```python
ruta="data/"
#ruta="/home/gari/Desktop/Jacobs_resourses/Advanced_project/latex/REPORT_1/data/"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import datetime
#from sklearn.cross_validation import train_test_split
from sklearn.neural_network import MLPRegressor
#Lectura de datos
dataset=pd.read_csv(ruta+'deepblue_bot/dump_history.txt',delimiter=";")
dataset['date'] = pd.to_datetime(dataset['time_stamp(17)'],unit='s')
dataset.set_index('date', inplace=True)

#for_plot=dataset['last(1)']['06/07/2018']

import seaborn as sns
font = {'family': 'serif',
        'weight': 'bold',
        'size': 16,
        }

#dataset_1_freq=dataset['5/27/2018':'6/07/2018']
dataset_d_freq=dataset['2018']
dataset_d_freq = dataset_d_freq.resample('D').mean()


#dataset_d_freq.shape

from scipy.interpolate import interp1d
import time


def interpolation_feactures(data,kind):
    col=data.columns
    array_df=[]
    contador=0
    for each_column in col:
        #print(each_column)
        data_for_analisis=data[each_column]
        data_for_analisis_no_na=data[each_column].dropna()
        new_index=data_for_analisis_no_na.index
        new_index=[ time.mktime(d_date.timetuple()) for d_date in  new_index]
        old_index=data_for_analisis.index
        old_index=[ time.mktime(d_date.timetuple()) for d_date in  old_index]
        
        #print(new_index)
        #data_for_mounth=data_for_analisis[fecha]
        function_interpolation = interp1d(new_index ,data_for_analisis_no_na, kind=kind)
        indeces=np.array(data_for_analisis.index)
        temp=pd.DataFrame(data=function_interpolation(old_index),
                          index=data_for_analisis.index, columns=[each_column])
        if contador==0:
            final=pd.DataFrame(data=function_interpolation(old_index),
                               index=data_for_analisis.index, columns=[each_column])
        else:
            final=final.join(temp)
        contador+=1
        #print(final)
    return final

data_set_interpolated=interpolation_feactures(dataset_d_freq,'cubic').drop('time_stamp(17)',axis=1)
data_set_interpolated=data_set_interpolated[['last(1)','sentiment_Twitter(15)','interest_trends(16)']]

TxGrowth=pd.read_csv(ruta+'ether_scan/export-TxGrowth.csv',delimiter=",")
TxGrowth['date'] = pd.to_datetime(TxGrowth['UnixTimeStamp'],unit='s')
TxGrowth.set_index('date', inplace=True)

```

```python

from sklearn.preprocessing import MinMaxScaler

scaler_aux = MinMaxScaler()

target_scale_auxiliar=scaler_aux.fit(data_set_interpolated)
#Isolate trends

#let us scale the data

from sklearn . preprocessing import StandardScaler,MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data_set_interpolated)

#Normalized data
scaled_data= scaler.transform(data_set_interpolated)



df_scaled_data=pd.DataFrame(scaled_data,
                            columns=data_set_interpolated.columns,index=data_set_interpolated.index)
```

```python
df_scaled_data=pd.DataFrame(scaled_data,
                            columns=data_set_interpolated.columns,index=data_set_interpolated.index)
df_scaled_data['lag_1']=df_scaled_data['last(1)'].shift(1)
df_scaled_data['target']=df_scaled_data['last(1)']-df_scaled_data['lag_1']
df_scaled_data['ajusted_target']=df_scaled_data['target'].shift(-1)
df_scaled_data=df_scaled_data.dropna()
prune_data=df_scaled_data
df_scaled_data
```

```python
#df_scaled_data
data_final=prune_data
X_data_frame=data_final[['sentiment_Twitter(15)','interest_trends(16)','Value']]
X=X_data_frame.values.reshape(len(X_data_frame),3)
y_data_frame=data_final[['ajusted_target']]
y=y_data_frame.values.reshape(len(y_data_frame),1)
```

```python
from sklearn.ensemble import GradientBoostingRegressor


from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')

# set the parameters
param_grid = [
        {
            'learning_rate':[0.01,0.02,0.03],
            'max_depth': [20,25,30,35],
            'n_estimators':[5,10,25,50]
        }
       ]

model=GradientBoostingRegressor(loss='lad',criterion='mae',random_state=42,tol=1e-8)
gs_clf = GridSearchCV(model,  param_grid, cv=2, scoring='neg_mean_absolute_error' )

gs_clf.fit(X, y)

print("Best parameters set found on the MLP classifier development set: ")
print(gs_clf.best_params_)
```

```python
df_y_pred=pd.DataFrame(gs_clf.predict(X),columns=['Prediction'],index=y_data_frame.index)

fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(16,7))
ax.plot(df_y_pred, '-o',  label="Hysteria Predicted on train data",color='blue')
ax.plot(y_data_frame, '-o',  label="Error  Actual on train",color='red')
plt.xticks(rotation=45)
ax.legend(loc='best')
#ax.set_xlabel(lab ,fontdict=font)
ax.set_ylabel(r'$hl_{t}$', fontdict=font)
#ax.set_title('Eth price one step forecast for 2019 ', fontdict=font)
#ax.xticks( rotation=45)
ax.grid(True)
```

```python

def sMAPE(y,h):
    h=h.reshape(len(h),1)
    y=y.reshape(len(y),1)
    #print('h',h.shape)
    #print('y',y.shape)
    num=np.abs(y-h)
    dem=(np.abs(y)+np.abs(h))/2
    ajust_dem=np.where(dem==0, 1, dem)
    sMAPE=num/ajust_dem
    return (np.sum(sMAPE)/len(y))*100
```

```python
y=metric['last(1)'].values
h=metric['error_and_lag1_predic'].values
sMAPE(y,h)
```
